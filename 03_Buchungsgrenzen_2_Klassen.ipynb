```python
from scipy.stats import norm
import numpy as np
```

```python
# Variablen
r1 = 300
r2 = 100
C = 160
mu1 = 35
sigma1 = 10
```

```python
#Schutzgrenze Klasse 1 (Littlewood) und Buchungsgrenze Klasse 2
G1 = np.round(norm(mu1, sigma1).ppf(1-r2/r1),0)  #ppf=percent point function (inverse of cumulative distribution function)
B2 = C-G1
print("Schutzgrenze Klasse 1: {}".format(G1))
print("Buchungsgrenze Klasse 2: {}".format(B2))
```

```python
# Gewinn
revenue = r1*G1 + r2*B2
print("Gewinn bei Auslastung der Klasse 1: {}".format(revenue))
```
